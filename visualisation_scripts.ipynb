```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r"C:\Users\24350\Desktop\EP\experiment2.xlsx")

df.columns = df.columns.str.strip()

df = df.rename(columns={"Step": "Step", "FFS": "FFS", "Response": "Response"})

df["Step"] = df["Step"].astype(str)

summary = df.groupby(["Step", "FFS"]).agg(
    LongHits=("Response", "sum"),
    TotalTrials=("Response", "count")
).reset_index()

summary["LongRate"] = summary["LongHits"] / summary["TotalTrials"]

print(summary)

plt.figure(figsize=(8, 5))
for ffs_type in summary["FFS"].unique():
    subset = summary[summary["FFS"] == ffs_type]
    plt.plot(subset["Step"], subset["LongRate"], marker="o", label=f"{ffs_type} FFS")

plt.title("Proportion of Judging the Vowel as 'Long'")
plt.xlabel("Step (1 = long, 2=mid, 3 = short)")
plt.ylabel("Proportion of 'Long' Responses")
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r"C:\Users\24350\Desktop\EP\experiment3.xlsx")

df.columns = df.columns.str.strip()

df = df.rename(columns={"Step": "Step", "Duration": "Duration", "Response": "Response"})

df["Step"] = df["Step"].astype(str)

summary = df.groupby(["Step", "Duration"]).agg(
    LongHits=("Response", "sum"),
    TotalTrials=("Response", "count")
).reset_index()

summary["LongRate"] = summary["LongHits"] / summary["TotalTrials"]

print(summary)

plt.figure(figsize=(8, 5))
for dur_type in summary["Duration"].unique():
    subset = summary[summary["Duration"] == dur_type]
    plt.plot(subset["Step"], subset["LongRate"], marker="o", label=f"{dur_type} Duration")

plt.title("Proportion of Judging the Vowel as 'Long'")
plt.xlabel("Step (1 = short spectrum, 2 = middle spectrum, 3 = long spectrum)")
plt.ylabel("Proportion of 'Long' Responses")
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
